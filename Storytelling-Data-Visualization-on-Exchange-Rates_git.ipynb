# Storytelling Data Visualization on Exchange Rates

The dataset we'll use for this project is available [here](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020).

The focus of this project will be on the exchange rate between the euro and the American dollar.

## Introducing the Dataset

```python
import pandas as pd

exchange_rates = pd.read_csv ("euro-daily-hist_1999_2020.csv")

exchange_rates.head()
```

```python
exchange_rates.info()
```

- There are `5699` rows and `41` columns in this table. From `4th January 1999` to `8th January 2021`, each column shows the currency rate for the nations indicated.

- The `Greek drachma`, `Slovenian tolar`, `Slovak koruna`, and `Cypriot pound` all have null values. With null values up to `50 percent`. 

- The majority of the columns are `strings`, although a few are `numeric`.

## Data Cleaning

The exchange rate between the `Euro` and the `American dollar` will be the project's main focus. As a result, we'll concentrate solely on them.


```python
# Rename `[US dollar]` and `Period\Unit:` to `US_dollar` and `Time`
exchange_rates.rename (columns={'[US dollar ]': 'US_dollar',
                                'Period\\Unit:': 'Time'},
                       inplace=True)

# Change `Time` to a `datetime` object
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

# Sort the values by `Time` in ascending order
exchange_rates.sort_values('Time', inplace=True)

# Reset the index ( and drop the initial index )
exchange_rates.reset_index(drop=True, inplace=True)

exchange_rates.head()
```

```python
# Isolate the related columns
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()

# Check the EUR-USD exchange rates values
euro_to_dollar['US_dollar'].value_counts()
```

- There are `62` inaccurate values (`-`) for the exchange rates, which resulted in the string data type for these columns.

- These `missing` values will be removed, and the columns will be converted to `float` data types.

```python
# Keep only the valid values
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']

# Convert `US_dollar` to `float` data type
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)

euro_to_dollar.info()
```

## Rolling Mean

We've finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.


```python
import matplotlib.pyplot as plt
%matplotlib inline 
# Enable Jupyter to display graphs

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.title('Euro to US Dollar - Daily Exchange Rate between 1999 to 2021')
plt.show()
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2,3,4,5,6], 
                           [7, 30, 50, 100, 365]):

    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold' )
        
plt.tight_layout()
plt.show()
```

We'll now calculate the rolling mean of the US dollar.

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

## Storying Telling Data Visualization Examples

### Financial Crisis Example


```python
financial_crisis_06_09 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year>=2006) & 
                                  (euro_to_dollar['Time'].dt.year<=2009)]

financial_crisis_07_08 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year>=2007) & 
                                  (euro_to_dollar['Time'].dt.year<=2008)]
```

```python
# Adding the FiveThirtyEight style
import matplotlib.style as style

# Setting FiveThirtyEight style
plt.style.use('fivethirtyeight')

# Setting size of our plot
fig, ax = plt.subplots(figsize=(8,4))

# plotting the graphs
ax.plot(financial_crisis_06_09['Time'], 
        financial_crisis_06_09['rolling_mean'],
        color= 'green', linewidth=1)

ax.plot(financial_crisis_07_08['Time'], 
        financial_crisis_07_08['rolling_mean'],
        color= 'red', linewidth=3)

# Highlighting the peak of the crisis
ax.axvspan(xmin=733122, xmax=733297, 
           ymin=0.085, 
           alpha=0.5, color='grey')

# Remove axes labels
ax.set_xlabel('')
ax.set_ylabel('')

# Adding separate tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

# Use axes.get_xticks to return the x ticks as a list of locations 
#ax.get_xticks()

# X-labels and changing label names
x = 732262
for year in [2006, 2007, 2008, 2009, 2010]:
    ax.text(x, 1.14, year, alpha=0.5, fontsize=11)
    x += 365

# Y-labels to only these
y = 1.193
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(732162, y, rate, alpha=0.5, fontsize=11)
    y += 0.1
    
# Adding a title 
ax.text(x=732162, y=1.625, 
        s="Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis", 
        fontsize=14, fontweight='semibold', color='#414141')

# Adding a subtitle 
ax.text(x=732162, y=1.6, 
        s='Euro-USD exhange rates between 2006 and 2010', 
        fontsize=12, color='#414141')    

# Adding a signature
ax.text(732162.0, 1.07, '©DATAQUEST' + ' '*94 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)
    
# Adding some transparency to the grid
ax.grid(alpha=0.5)
    
plt.show()    
```
